```python
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_size = 1000

salary = np.random.randint(10000,50000,data_size)
experience = np.random.randint(0,15,data_size)
front_end = [bool(random.getrandbits(1)) for x in range(data_size)]
back_end = [bool(random.getrandbits(1)) for x in range(data_size)]

# print(salary,experience,front_end,back_end)
```

```python
data = np.empty((data_size,4),dtype='O')
ind = 0
for elem in [experience,front_end,back_end,salary]:
    data[:,ind] = elem
    ind+=1
```

```python
df = pd.DataFrame(data = data,columns=['experience','front_end','back_end','salary'])
```

```python
fig,ax = plt.subplots(1,3)
fig.set_figheight(5)
fig.set_figwidth(15)
ax[0].plot(df.experience)
ax[1].plot(df.front_end)
ax[2].plot(df.back_end)
plt.show()
```

```python
# as my value is random this plot does not make any sense
fig,ax = plt.subplots(1,3)

ax[0].plot(df.salary,df.experience)
ax[1].plot(df.front_end,df.experience)
ax[2].plot(df.back_end,df.experience)
plt.ylabel("Experience")
plt.show()
```

```python
back_end_programmer = df.back_end[df.back_end==True].count()
front_end_programmer = df.front_end[df.front_end==True].count()
print(back_end_programmer,front_end_programmer)
plt.pie([back_end_programmer,front_end_programmer],labels=['back_end','front_end'],autopct='%1.2f%%')
```
